```python
import cv2
from ultralytics import YOLO

model = YOLO('yolov8n.pt')
model.train(data='./dataset/data.yaml', epochs=50, device=0)
```

```python
model = YOLO('runs/detect/train2/weights/best.pt')

video = cv2.VideoCapture('./test_video/video_2023-06-23_12-33-40.mp4')
cv2.namedWindow('Video', cv2.WINDOW_NORMAL)

playing = True
while True:
    if playing:
        ret, frame = video.read()
        if ret:
            results = model(frame)
            for box in results[0].boxes:
                if(box.conf[0].item() > 0.5):
                    x1, y1, x2, y2 = box.xyxy[0]
                    name = results[0].names[box.cls[0].item()]
                    cv2.putText(frame, f'{name}', (int(x1), int(y1 - 10)), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 0), 2)
                    if(box.cls[0].item() == 0):
                        color = (0, 255, 0)
                    elif(box.cls[0].item() == 1):
                        color = (0, 0, 255)
                    elif(box.cls[0].item() == 2):
                        color = (0, 255, 255)
                    cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), color, 5)
            cv2.imshow('Video', frame)
        else:
            video.set(cv2.CAP_PROP_POS_FRAMES, 0)
    key = cv2.waitKey(1)
    if key == 27 or cv2.getWindowProperty("Video", cv2.WND_PROP_VISIBLE) < 1:
        break
    elif key == ord('p'):
        playing = not playing

video.release()
cv2.destroyAllWindows()
```
